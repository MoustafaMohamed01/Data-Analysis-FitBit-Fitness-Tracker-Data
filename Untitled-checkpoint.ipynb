```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import datetime as dt
```

```python
df = pd.read_csv('dailyActivity_merged.csv')
df.head()
```

```python
cols = ['Id' , 'ActivityDate' , 'TotalSteps' , 'VeryActiveMinutes' , 'FairlyActiveMinutes' , 'LightlyActiveMinutes' , 'SedentaryMinutes' , 'Calories']
```

```python
df2 = df[cols]
df2.head()
```

```python
df2.rename(columns = {'ActivityDate' : 'date'} , inplace = True)
df2.head()
```

```python
df2['TotalMinutes'] = df2.VeryActiveMinutes + df2.FairlyActiveMinutes + df2.LightlyActiveMinutes + df2.SedentaryMinutes
```

```python
df2.head()
```

## Create Total Hours Column 

```python
df2['TotalHours'] =round(df2.TotalMinutes / 60 , 2)
df2.head()
```

```python
df2.info()
```

## Convert date Column from object to datetime64

```python
df2.date = pd.to_datetime(df2.date)
df2.head()
```

```python
df2.info()
```

## Day Of Week Column

```python
df2['DayOfWeek'] = df2.date.dt.day_name()
df2.head()
```
